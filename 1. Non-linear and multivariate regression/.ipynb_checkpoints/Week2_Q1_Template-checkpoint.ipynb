```python
import numpy as np
from matplotlib import pyplot as plt

class nlr:
    def grad(self,x,y,w):
        J = 0 
        x = np.transpose(x)
        p = w@x-y
        x= np.transpose(x)
        return p@x/len(x)

    def computeCost(self,x,y,w):
        J = 0 
        x = np.transpose(x)
        p = w@x-y
        x= np.transpose(x)
        z=p*p
        return np.sum(z)/(2*len(x))
    
    def BgradientDescent(self, x, y, w1, alpha, iters):
        m =  len(x)  
        p = len(w1)
        w = np.array([0.0]*p) 
        
        J_history = []

        for i in range(iters):
            temp = self.grad(x, y, w)
            for j in range(p):
                w[j]-= alpha*temp[j]
            J_history.append(self.computeCost(x, y, w))
        return w, J_history
  
    def SgradientDescent(self, x, y, w1, alpha, iters):
        m =  len(x)
        p = len(w1)
        w = np.array([0.0]*p)
        J_history_s = []   # Using a python list to save cost in every iteration
        n= int(m/3)
        n = 100 if n>100 else n
        for i in range(iters):
            r = np.random.randint(m)
            for k in range(p):
                J=0
                for i in range(r,r+n):
                    temp = 0
                    for j in range(p):
                        temp+= w[j]*x[(i)%m][j]
                    temp-= y[(i)%m]
                    J+= temp*x[(i)%m][k]
                w[k] -= alpha*J/n
            J_history_s.append(self.computeCost(x, y, w))
        return w, J_history_s
  
  # This function implements line search Secant method
    def ls_secant(self,x,y,w,d):
    # d is search direction d = -grad(J). Refer class and Lab notes
        epsilon = 10**(-4) # Line search tolerance

        alpha_curr = 0.0     # Alpha (x_i-1)
        alpha = 0.01       # initial value (x_i)

        dphi_zero =  0.0       # dphi_zero = (d^T)(grad J(w_0) # At every alpha updation loop you will have a given initial weight vector (w_0)
        for i in range(3):
            dphi_zero -= d[i]*d[i]
        dphi_curr = dphi_zero  # required for first alpha iteration
        while abs(dphi_curr) > (epsilon*abs(dphi_zero)):  # tolerance or looping criteria used here
            w_temp = [w[0]+alpha*d[0], w[1]+alpha*d[1], w[2]+alpha*d[2]]
            dphi_temp = 0.0
            grad_J_temp = self.grad(x, y, w_temp)
            for i in range(3):
                dphi_temp += d[i]*grad_J_temp[i]
            alpha_temp = (dphi_temp*alpha_curr - dphi_curr*alpha)/(dphi_temp-dphi_curr)
            alpha_curr = alpha
            alpha = alpha_temp
            dphi_curr = dphi_temp
        # write loop to update alpha
        

        return alpha

  # This function optimizes the weights w_0, w_1, w_2. Batch Gradient Descent method using variable alpha which you previously updated using ls_secant() method
    def AgradientDescent(self,x, y, w1, iters):
        m =  len(x)  # number of training examples
        w = np.array([0.0]*len(w1)) # To keep a copy of original weights
        eps = 10**(-12); # tolerance for J_history

        J_history_a = [0]

        for i in range(iters):
            d = - self.grad(x, y, w)
            alpha = self.ls_secant(x, y, w, d)
            for j in range(3):
                w[j]+= alpha*d[j]
            J_history_a.append(self.computeCost(x, y, w))
            
            if i>0 and (J_history_a[i] -J_history_a[i+1] < eps):
                print('No. of iterations',i)
                break
        return w, J_history_a
```

### Data from .txt file

```python
data = np.loadtxt('prob1data.txt', delimiter=',')
[a, b] = data
plt.scatter(a,b)
plt.xlabel('Time')
plt.ylabel('Height')
plt.title('DATA')
```

```python
a = a.reshape(-1, 1) 
x = np.concatenate((np.ones(a.shape), a, a*a), axis = 1)
y = b

n = nlr()
Bvec, Bcost = n.BgradientDescent(x, y, [0.0, 0.0, 0.0], 0.01, 50000)
```

```python
Svec, Scost = n.SgradientDescent(x, y, [0.0, 0.0, 0.0], 0.01, 50000)
```

```python
p1 = plt.subplots()[1]
p1.plot(range(50000), Bcost)
p1.title.set_text('BATCH GRADIENT DESCENT')
p1.set_xlabel('Iteration')
p1.set_ylabel('Cost')

p2 = plt.subplots()[1]
p2.plot(range(50000), Scost)
p2.title.set_text('STOCHASTIC GRADIENT DESCENT')
p2.set_xlabel('Iteration')
p2.set_ylabel('Cost')
```
